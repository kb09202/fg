```python
from sklearn.ensemble import IsolationForest
import numpy as np
import matplotlib.pyplot as plt

# Donn√©es
data = np.random.rand(100, 2)

# Mod√®le
model = IsolationForest(contamination=0.1)
model.fit(data)
outliers = model.predict(data)

# Visualisation
plt.scatter(data[:, 0], data[:, 1], c=outliers, cmap="coolwarm", edgecolor="k")
plt.title("D√©tection d'Anomalies (Isolation Forest)")
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Donn√©es
iris = load_iris()
X, y = iris.data, iris.target

# Mod√®le
model = DecisionTreeClassifier()
model.fit(X, y)

# Visualisation
plt.figure(figsize=(12, 8))
plot_tree(model, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.title("Arbre de D√©cision")
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Donn√©es
iris = load_iris()
X = iris.data
y = iris.target

# PCA
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

# Visualisation
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap="viridis")
plt.title("R√©duction de Dimension (PCA)")
plt.xlabel("Composante 1")
plt.ylabel("Composante 2")
plt.colorbar(label="Classe")
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import ConfusionMatrixDisplay

# Donn√©es
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)

# Mod√®le
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)

# √âvaluation
y_pred = model.predict(X_test)

# Visualisation de la matrice de confusion
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
plt.title("Matrice de Confusion")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Donn√©es synth√©tiques
data = np.random.rand(100, 2)

# Clustering
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)
labels = kmeans.labels_

# Visualisation
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap="viridis", label="Points")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color="red", marker="X", label="Centroides")
plt.title("K-Means Clustering")
plt.legend()
plt.show()
```
